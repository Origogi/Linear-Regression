# Example 1 - A simple cost function

## $$f(x) = x^2 + x +1$$

```python
def f(x) :
    return x**2 + x + 1
```

```python
# Make Data
x_1 = np.linspace(start=-3, stop=3, num=10)
```

```python
print(x_1)
```

```python
# Plot
plt.xlim([-3,3])
plt.ylim(0,8)
plt.xlabel('X', fontsize=16)
plt.ylabel('f(x)', fontsize=16)

plt.plot(x_1, f(x_1))
```
